```python
# %pip install pyreadr --quiet
# %pip install requests --quiet
import pyreadr
import requests
```

```python
# Raw link to data
url = "https://raw.githubusercontent.com/DataSlingers/clustRviz/master/data/authors.rda"

# Download the file
with open("authors.rda", "wb") as f:
    f.write(requests.get(url).content)

# Load the .rda file
results = pyreadr.read_r("authors.rda")
results = dict(results)
# results.keys() # keys = [authors]
df = results['authors']
df = df.reset_index(drop = False).rename(columns = {'index':'BookID'})
df
```

```python
book_id = df['BookID']
book_id.value_counts()
# 4 different books ; w/ 317 - Austen, ..., 55 - Milton.
```

Data explanation - a row represents a book chapter with each column representing the word counts of key words in that chapter.  We have a BookID feature that tells us which book this corresponds to.  
For the purposes of this project we will only use this BookID column for validation of our techniques.
